# Clustering

## Annual Income and Spending Score

```python
df_income_score = df.iloc[:,[3,4]]
```

```python
df_income_score.describe()
```

##### Elbow Method

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans


kmeans = KMeans(random_state=0)
visualizer = KElbowVisualizer(kmeans, k=(1,11))

visualizer.fit(df_income_score)
visualizer.show()
```

```python
from yellowbrick.cluster import SilhouetteVisualizer

plt.figure(figsize=(2 * 5,  10 * 4))

scores = {}
for n_clusters in range(2, 8):
    plt.subplot(10, 2, n_clusters - 1)
    kmeans = KMeans(n_clusters, random_state=42)
    visualizer = SilhouetteVisualizer(kmeans, colors='yellowbrick')
    visualizer.fit(df_income_score)
    scores[n_clusters] = visualizer.silhouette_score_
    plt.title(f'clusters: {n_clusters} score: {visualizer.silhouette_score_}')
```

```python
sorted(scores.items(), key=lambda kv: kv[1], reverse=True)
```

```python
kmeans = KMeans(n_clusters=5, random_state=0)
kmeans.fit(df_income_score)
identified_clusters= kmeans.fit_predict(df_income_score)
```

```python
data_with_clusters = df.copy()
data_with_clusters['Label'] = identified_clusters 
sns.scatterplot(x = 'Annual Income (k$)',y = 'Spending Score (1-100)',hue="Label",  
                 palette="rainbow", legend='full',data = data_with_clusters  ,s = 60 )
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)') 
plt.title('Spending Score vs Annual Income', fontsize = 15)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], 
            c='black', edgecolor='black', alpha=0.7, s=100)
plt.show()
```

```python
all_clusters = [data_with_clusters[data_with_clusters['Label'] == i] for i in range(5)]
```

```python
all_clusters[1].describe()
```

```python
plt.figure(figsize=(2 * 5,  10 * 4))

scores = {}
for n_clusters in range(2, 8):
    plt.subplot(10, 2, n_clusters - 1)
    kmeans = KMeans(n_clusters, random_state=42)
    visualizer = SilhouetteVisualizer(kmeans, colors='yellowbrick')
    visualizer.fit(df1)
    scores[n_clusters] = visualizer.silhouette_score_
    plt.title(f'clusters: {n_clusters} score: {visualizer.silhouette_score_}')
```

```python
data_with_clusters = df1.copy()
data_with_clusters['label'] = identified_clusters 

from mpl_toolkits import mplot3d
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data_with_clusters['Age'][data_with_clusters.label == 0], data_with_clusters["Annual Income (k$)"][data_with_clusters.label == 0], data_with_clusters["Spending Score (1-100)"][data_with_clusters.label == 0], c='blue', s=60)
ax.scatter(data_with_clusters['Age'][data_with_clusters.label == 1], data_with_clusters["Annual Income (k$)"][data_with_clusters.label == 1], data_with_clusters["Spending Score (1-100)"][data_with_clusters.label == 1], c='red', s=60)
ax.scatter(data_with_clusters['Age'][data_with_clusters.label == 2], data_with_clusters["Annual Income (k$)"][data_with_clusters.label == 2], data_with_clusters["Spending Score (1-100)"][data_with_clusters.label == 2], c='green', s=60)
ax.scatter(data_with_clusters['Age'][data_with_clusters.label == 3], data_with_clusters["Annual Income (k$)"][data_with_clusters.label == 3], data_with_clusters["Spending Score (1-100)"][data_with_clusters.label == 3], c='orange', s=60)
ax.scatter(data_with_clusters['Age'][data_with_clusters.label == 4], data_with_clusters["Annual Income (k$)"][data_with_clusters.label == 4], data_with_clusters["Spending Score (1-100)"][data_with_clusters.label == 4], c='purple', s=60)
ax.view_init(30, 185)
plt.title('Spending Score vs Age vs Annual Income', fontsize = 15)
plt.xlabel("Age", fontsize = 13)
plt.ylabel("Annual Income (k$)", fontsize = 13)
ax.set_zlabel('Spending Score (1-100)', fontsize = 13)
plt.show()

```
